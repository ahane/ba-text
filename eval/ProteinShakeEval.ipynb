```python
def df_from_radio(json_dict):
    print type(json_dict)
    distances = [tuple(x) for x in json_dict['connections']]
    distdf = DataFrame(distances, columns=['edge', 'distance'])
    dist_with_index = distdf.set_index('edge')
    return dist_with_index
```

```python
def df_from_ProteinShake(json_dict, name):
    predictions = [x.items()[0] for x in json_dict['predictions']]
    preddf = DataFrame(predictions, columns=['edge', name])
    pred_with_index = preddf.set_index('edge')
    return pred_with_index
```

```python
def df_with_tp_rates(df, name, threshold=8):
    df_copy = df[['edge', 'distance', name]].copy()
    all_predicted = df_copy[df_copy[name] > -1]
    all_predicted_and_measured = all_predicted[all_predicted['distance'] > -1]
    all_predicted['dist_under_thres'] = all_predicted['distance'].apply(lambda x: x <= threshold and x > -1)
    sorted_by = all_predicted.sort(name, ascending=False)
    
    sorted_by['positives'] = sorted_by['dist_under_thres'].apply(lambda x: 1 if x else 0)
    sorted_by['cuml_positives'] = sorted_by['positives'].cumsum(axis=0)
    
    sorted_by['count_helper'] = 1
    sorted_by['predictions_count'] = sorted_by['count_helper'].cumsum(axis=0)
    del sorted_by['count_helper']
    
    sorted_by['tp_rate'] = sorted_by['cuml_positives'] / sorted_by['predictions_count']
    
    return sorted_by
```

```python
def df_with_tp_rates_reverse(df, name, threshold=8):
    df_copy = df[['edge', 'distance', name]].copy()
    all_predicted = df_copy[df_copy[name] > 0]
    all_predicted_and_measured = all_predicted[all_predicted['distance'] > -1]
    all_predicted['dist_under_thres'] = all_predicted['distance'].apply(lambda x: x <= threshold and x > -1)
    sorted_by = all_predicted.sort(name, ascending=True)
    
    sorted_by['positives'] = sorted_by['dist_under_thres'].apply(lambda x: 1 if x else 0)
    sorted_by['cuml_positives'] = sorted_by['positives'].cumsum(axis=0)
    
    sorted_by['count_helper'] = 1
    sorted_by['predictions_count'] = sorted_by['count_helper'].cumsum(axis=0)
    del sorted_by['count_helper']
    
    sorted_by['tp_rate'] = sorted_by['cuml_positives'] / sorted_by['predictions_count']
    
    return sorted_by
```

```python
def load_dfs(dir_path, path_to_radio_json):
    dfs = []
    
    with open(path_to_radio_json) as f:
        radiojson = json.loads(f.read())
    print type(radiojson)
    dfs.append(df_from_radio(radiojson))
    
    files = [x[:-5] for x in os.listdir(dir_path)]
    for filename in files:
        with open(dir_path+filename+'.json') as f:
            file_json = json.loads(f.read())
            dfs.append(df_from_ProteinShake(file_json, filename))
    
    alldf = pd.concat(dfs, axis=1)
    alldf['edge'] = alldf.index
    alldf_dropped = alldf.dropna()
    return alldf_dropped
```

```python
def load_dfs_tprate(dir_path, path_to_radio_json):
    names = [x[:-5] for x in os.listdir(dir_path)]
    alldf_tp = load_dfs(dir_path, path_to_radio_json)
    tp_dfs_names = [(df_with_tp_rates(alldf_tp, name), name) for name in names]
    return tp_dfs_names
```

```python
def load_dfs_tprate_reverse(dir_path, path_to_radio_json):
    names = [x[:-5] for x in os.listdir(dir_path)]
    alldf_tp = load_dfs(dir_path, path_to_radio_json)
    tp_dfs_names = [(df_with_tp_rates_reverse(alldf_tp, name), name) for name in names]
    return tp_dfs_names
```

```python
def parse_filename(filename):
    split = filename.split('_')
    if len(split) == 6: #new naming
        l1, l2, lr, iterations, steps, delta = split
        l1 = l1.replace('-', '.', 1)[2:]
        #l2 = l2.replace('-', '.', 1)[2:]
        lr = lr.replace('-', '.', 1)[2:]
        #return int(iterations), int(steps)
    elif len(split) == 3: #oldnameing
        match = re.match('l1(\d\-[\d|E|\-]+)l2(\d\-[\d|E|\-]+)lr(\d\-[\d|E|\-]+)\_(\d)\_(\d)', filename)
        l1 = match.group(1).replace('-', '.', 1)
        l2 = match.group(2).replace('-', '.', 1)
        lr = match.group(3).replace('-', '.', 1)
    else:
        print('could read name')
        l1 = 'error'
        lr = 'error'
    return l1, lr
```

```python
def add_hyperparameters(df_with_tp, filename):
    df_hyper = df_with_tp[['predictions_count', 'tp_rate']].copy()

    l1, lr = parse_filename(filename)
    df_hyper['l1'] = l1
    df_hyper['lr'] =  lr

    df_hyper = df_hyper.reset_index()
    del df_hyper['index']
    
    return df_hyper
    
```

```python
def plot_multiple(dfs):

    fig = plt.figure()
    ax = fig.add_subplot(111)
    num_contacts = dfs[0][0]['positives'].sum()
    for df, name in dfs:
        if name != 'MI':
            l1, lr = parse_filename(name)
            name = 'L1: '+l1+', LR: '+lr
        plt.plot(df['predictions_count'], df['tp_rate'], label=name)
        plt.xscale('log')
        
        final_value = round(df['tp_rate'][num_contacts], 4)
        ax.annotate(final_value, xy=(num_contacts, final_value))
    
    plt.xlabel("Number of Contacts")
    plt.xlim(1, num_contacts)
    plt.ylabel("TP-Rate until n-th Prediction")
    plt.legend()
    plt.savefig('eps/'+CURRENT_PF+"_"+dfs[0][1]+'.eps')
```
