```python
import json
import pandas as pd
from matplotlib import pyplot as plt
from pandas import DataFrame, Series
import seaborn as sns
from IPython.display import SVG
%pylab inline
```

```python
def df_from_ProteinShake(json_dict, name):
    predictions = [x.items()[0] for x in json_dict['predictions']]
    preddf = DataFrame(predictions, columns=['edge', name])
    pred_with_index = preddf.set_index('edge')
    return pred_with_index
```

```python
with open('25/MI.json') as f:
    mi14 = df_from_ProteinShake(json.loads(f.read()), 'mi')
```

```python
with open('other/PF00017/MI.json') as f:
    mi17 = df_from_ProteinShake(json.loads(f.read()), 'mi')
```

```python
with open('other/PF00035/MI.json') as f:
    mi35 = df_from_ProteinShake(json.loads(f.read()), 'mi')
```

```python
all_mi = [mi14, mi17, mi35]
for df in all_mi:
    df.reset_index(inplace=True)
    del df['edge']
```
